```python
import requests
from bs4 import BeautifulSoup
import lxml

url = 'http://rottentomatoes.com'
response = requests.get(url)

html = response.text
soup = BeautifulSoup(html,'lxml')
trs = soup.find('table', attrs={'id':'Top-Box-Office'}).find_all('tr')

box_office_list = []
for tr in trs:
    score = int(tr.find('span',attrs={'class':'tMeterScore'}).text[:-1])
    title = tr.find('td',attrs={'class':'middle_col'}).text.strip('\n')
    revenue = float(tr.find('td',attrs={'class':'right_col'}).text.strip('\n$M'))
    link = url + tr.find('a').attrs['href']
    box_office_list.append((
        score, 
        title, 
        revenue,
        link,
    ))
box_office_list


```
