```python
'''
To do ::: To extract the information from top box office on rottentomato.com
1. score
2. movie url
3. movie title
4. movie revenue
'''

# import requsts and bs4
import requests
from bs4 import BeautifulSoup


# request html to rottentomatoes.com
url = 'https://www.rottentomatoes.com'
response = requests.get(url)

# html parsing
# find the table where 'id' is 'Top-Box-Office'
html = response.text
soup = BeautifulSoup(html,'html.parser')
trs = soup.find('table', attrs={'id':'Top-Box-Office'}).find_all('tr')


movie_list = []
with open('rottentomato_top_box_office_info.csv','w') as f:
    for tr in trs:
        movie_score = tr.find('span',attrs={'class':'tMeterScore'}).text[:-1]
        movie_link = url + tr.find('a').attrs['href']
        movie_title = tr.find('td',attrs={'class':'middle_col'}).text[1:-1]
        movie_revenue = tr.find('td',attrs={'class':'right_col right'}).text[1:-1]
        movie_list.append((movie_score,movie_link, movie_title, movie_revenue))
        f.write("{score},{link},{title},{revenue}\n".format(
               score = movie_score,
               link = movie_link,
               title = movie_title,
               revenue = movie_revenue))

print(movie_list)
```
